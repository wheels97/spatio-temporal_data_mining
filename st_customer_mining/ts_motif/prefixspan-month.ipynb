```python
from prefixspan import PrefixSpan
import pandas as pd
import numpy as np
```

# 读入已转码POI数据

```python
from sklearn.externals import joblib
scan = joblib.load('newPOI-scan.pydata')
```

```python
scan_s = scan[['open_id','crt_date','crt_time','type']].dropna(axis=0,how='any')
# scan_s = scan[['open_id','crt_date','crt_time','type']].fillna(0)
scan_s = scan_s.reset_index()
```

```python
scan_s = scan_s[['open_id','crt_date','crt_time','type']]
scan_s
```

# POI转换代号
代号对照表:
 a 其他
 b 学校
 c 住宅
 d 公司
 e 娱乐场所\商场
 f 交通设施
 g 生活服务\汽车相关
 h 餐饮行业
 i 休闲
 j 风景名胜
 k 政府机构
 l 地名
 m 医疗

```python
type = scan_s['type'].values
mtype = []
for i in range(len(type)):
    if '校' in str(type[i]):
        mtype.append('B')
    elif '宅' in str(type[i]):
        mtype.append('C')
    elif '公司' in str(type[i]):
        mtype.append('D')    
    elif '购物' in str(type[i]):
        mtype.append('E')
    elif '交通' in str(type[i]):
        mtype.append('F')
    elif '生活' in str(type[i]) or '汽车' in str(type[i]) or '摩托车' in str(type[i]) or '住宿' in str(type[i]) or '金融'in str(type[i]):
        mtype.append('G')
    elif '餐饮' in str(type[i]):
        mtype.append('H')
    elif '休闲' in str(type[i]):
        mtype.append('I')
    elif '医疗' in str(type[i]):
        mtype.append('M')
    elif '风景' in str(type[i]):
        mtype.append('J')
    elif '政府' in str(type[i]):
        mtype.append('K')
    elif '地名' in str(type[i]):
        mtype.append('L')
    else:
        mtype.append('A')
```

```python
df_mtype = pd.DataFrame(mtype)
df_mtype[0].value_counts()
```

```python
len(scan_s)
```

```python
scan_s['mtype'] = df_mtype
```

```python
scan_s
```

```python
df_1 = scan_s.sort_values(by = ['open_id','crt_date','crt_time']).reset_index()
df_1
```

```python
numb = df_1.groupby('open_id')
df_1a = numb.filter(lambda x: len(x) > 5) .reset_index()
```

```python
df_1a
```

# crt_time转换为时间段标签

```python
time = df_1a['crt_time'].values
time
```

```python
timeline = []
for i in range(len(time)):
    hour = int(str(time[i])[:-4])
    if hour >= 6 and hour < 12:
        timeline.append('mor')
    elif hour >= 12 and hour < 19 :
        timeline.append('noon')
    elif hour >= 19 and hour < 24 :
        timeline.append('eve')
    elif hour == 24 or hour <6:
        timeline.append('night')
df_time = pd.DataFrame(timeline)
df_1a['mtime'] = df_time
df_1a
```

```python
gtime = df_1a['crt_date'].values
gmonth = []
for i in range(len(gtime)):
    gmonth.append(str(gtime[i])[2:6])
```

```python
df_1a['month'] = pd.DataFrame(gmonth)
df_1a
```

# 生成用户序列

```python
df_id = df_1a[['open_id','mtype','mtime','month']]
```

```python
df_id['mitem'] = df_id['mtype']+ df_id['mtime']
```

```python
df_id
```

```python
dflist = df_id.values
```

```python
namelist = []
typelist = []
typelist_1 = []
mixlist = []
initname = 0
nowmonth = 0
nameindex = -1
for i in range(len(dflist)):
    if dflist[i][0] == initname and dflist[i][3] == nowmonth:
        if dflist[i][1] != typelist[nameindex][-1]: 
            typelist[nameindex].append(dflist[i][1])
        typelist_1[nameindex].append(dflist[i][1])
        mixlist[nameindex].append(dflist[i][4])
    else:
        initname = dflist[i][0]
        nowmonth = dflist[i][3]
        nameindex += 1
        namelist.append(initname+'_'+nowmonth)
        typelist.append([dflist[i][1]])
        typelist_1.append([dflist[i][1]])
        mixlist.append([dflist[i][4]])
```

```python
len(namelist)
```

```python
typelist
```

生成了namelist,typelist,mixlist

```python
db = typelist_1
ps = PrefixSpan(db)
ps.minlen = 3
ps.maxlen = 5
print(ps.topk(5))
```

```python
db = typelist
ps = PrefixSpan(db)
ps.minlen = 3
ps.maxlen = 5
print(ps.topk(15))
```

```python
db = mixlist
ps = PrefixSpan(db)
ps.minlen = 3
ps.maxlen = 5
print(ps.topk(5))
```
